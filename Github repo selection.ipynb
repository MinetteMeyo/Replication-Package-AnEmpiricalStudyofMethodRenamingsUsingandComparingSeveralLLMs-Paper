```python
import requests

def get_real_world_repos(language, min_stars=1000, min_forks=500, min_size=500):
    """Fetch real-world repos based on stars, forks, and size."""
    GITHUB_API_URL = "https://api.github.com/search/repositories"
    query = f"language:{language} stars:>{min_stars} forks:>{min_forks} size:>{min_size} pushed:>2024-01-01"
    
    response = requests.get(GITHUB_API_URL, params={"q": query, "per_page": 11})
    
    if response.status_code != 200:
        print(f"Error fetching {language} repos:", response.json())
        return []
    
    repos = response.json()["items"]
    
    # Additional filtering: Ignore tutorials & docs
    filtered_repos = []
    for repo in repos:
        repo_name = repo["full_name"]
        repo_url = repo["html_url"]
        repo_size = repo["size"]
        repo_desc = repo["description"] or ""
        
        # Skip tutorial/documentation-style repos
        if any(keyword in repo_desc.lower() for keyword in ["tutorial", "example", "awesome", "documentation", "docs", "guide"]):
            continue
        
        filtered_repos.append((repo_name, repo_url, repo_size))
    
    return filtered_repos

# Fetch repos for Java, Python, and JavaScript
languages = ["Java", "Python", "JavaScript"]

for lang in languages:
    repos = get_real_world_repos(lang)
    print(f"\n🔹 Top {lang} Repos:")
    for name, url, size in repos:
        print(f"  - {name}: {url} (Size: {size} KB)")
```
